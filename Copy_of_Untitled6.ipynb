```python
# 1. Create a dictionary with 5 products as keys and their stock quantities as values
products = {
    'Laptop': 15,
    'Mouse': 50,
    'Keyboard': 20,
    'Monitor': 10,
    'Webcam': 5
}
print("Initial Inventory:")
print(products)
print("\n")

# 2. Add a new product and update the quantity of an existing one
products['Headphones'] = 30  # Add a new product
products['Mouse'] = 45       # Update existing product quantity
print("Inventory after adding new product and updating an existing one:")
print(products)
print("\n")

# 3. Write a function to return products with stock less than 10
def get_low_stock_products(inventory_dict):
    low_stock = {}
    for product, quantity in inventory_dict.items():
        if quantity < 10:
            low_stock[product] = quantity
    return low_stock

low_stock_items = get_low_stock_products(products)
print("Products with stock less than 10:")
print(low_stock_items)
print("\n")

# 4. Delete a discontinued product and display the updated dictionary
del products['Webcam'] # Delete the discontinued product 'Webcam'
print("Inventory after deleting 'Webcam':")
print(products)
print("\n")

# 5. Use .items() to loop through and print each product with its quantity
print("Current Product Inventory (using .items()):")
for product, quantity in products.items():
    print(f"{product}: {quantity}")
```

### Exercise 2: Lists in Logistics Management (Supply Chain Sector)

**Topic:** Python Lists (Marks: 20)

**Scenario:** A logistics company tracks delivery routes and package counts.

**Task:**
1.  Create a list of 10 delivery routes, each represented as a string. (4 marks)
2.  Append a new route and remove a discontinued one. (4 marks)
3.  Sort the list alphabetically and reverse it. (4 marks)
4.  Count how many routes start with the letter “N”. (4 marks)
5.  Use list comprehension to create a new list of routes longer than 15 characters. (4 marks)

```python
# 1. Create a list of 10 delivery routes, each represented as a string
routes = [
    'NorthStar Transport',
    'SouthCargo Route',
    'EastFlow Movers',
    'WestBridge Logistics',
    'CentralHub Logistics',
    'SkyTrack Freight',
    'RapidShift Delivery',
    'Oceanic Express',
    'National Carrier Path',
    'Global Connect'
]
print("Initial Delivery Routes:")
print(routes)
print("\n")

# 2. Append a new route and remove a discontinued one
routes.append('MegaRoute Express')  # Append a new route
routes.remove('Oceanic Express')     # Remove a discontinued route
print("Routes after adding and removing:")
print(routes)
print("\n")

# 3. Sort the list alphabetically and reverse it
routes.sort() # Sort alphabetically
routes.reverse() # Reverse the sorted list
print("Routes sorted alphabetically and then reversed:")
print(routes)
print("\n")

# 4. Count how many routes start with the letter “N”
count_N = 0
for route in routes:
    if route.startswith('N'):
        count_N += 1
print(f"Number of routes starting with 'N': {count_N}")
print("\n")

# 5. Use list comprehension to create a new list of routes longer than 15 characters
long_routes = [route for route in routes if len(route) > 15]
print("Routes longer than 15 characters (using list comprehension):")
print(long_routes)
```

### Explanation: Why Tuples are Suitable for Storing Patient Vitals (Task 3)

Tuples are ideal for storing patient vitals because:

*   **Immutability:** Once created, the elements of a tuple cannot be changed. This is crucial for medical records where data integrity is paramount. You wouldn't want a patient's recorded vitals to be accidentally altered.
*   **Data Integrity:** The immutability ensures that a specific set of vitals (like blood pressure and heart rate taken at a particular time) remains constant as an unchangeable record.
*   **Ordered Collection:** Tuples maintain the order of elements, which is important for consistently representing vital signs (e.g., always knowing the heart rate is the fourth element).
*   **Heterogeneous Data:** Tuples can store different data types (e.g., string for name, integer for age, float for blood pressure), which is necessary for patient vital records.
*   **Read-Only Access:** They are optimized for read-only operations, making them efficient for retrieving historical vital records.

```python
# 1. Create a tuple for a patient containing: (Name, Age, BloodPressure, HeartRate)
patient_vitals = ("Alice Smith", 34, "120/80", 72)
print(f"Patient Vitals Tuple: {patient_vitals}")
print("\n")

# 2. Access and print the patient's age and heart rate
patient_age = patient_vitals[1]    # Age is at index 1
patient_heart_rate = patient_vitals[3] # HeartRate is at index 3
print(f"Patient's Age: {patient_age}")
print(f"Patient's Heart Rate: {patient_heart_rate}")
print("\n")

# 4. Convert the tuple to a list, update the heart rate, and convert it back to a tuple
print(f"Original Heart Rate: {patient_vitals[3]}")
patient_vitals_list = list(patient_vitals)
patient_vitals_list[3] = 76 # Update heart rate
patient_vitals_updated = tuple(patient_vitals_list)
print(f"Updated Patient Vitals Tuple (after heart rate change): {patient_vitals_updated}")
print("\n")

# 5. Create a tuple of 5 patients and write code to extract all names
all_patients_vitals = (
    ("Alice Smith", 34, "120/80", 72),
    ("Bob Johnson", 52, "130/85", 78),
    ("Charlie Brown", 28, "115/75", 68),
    ("Diana Prince", 41, "125/82", 75),
    ("Eve Adams", 60, "140/90", 80)
)

print("All Patient Vitals (Tuple of Tuples):")
for patient in all_patients_vitals:
    print(patient)
print("\n")

patient_names = [patient[0] for patient in all_patients_vitals]
print(f"Extracted Patient Names: {patient_names}")
```
